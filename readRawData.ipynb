```python
import os
dir_path = '../rawdata'
flist = os.listdir(dir_path)
print(flist)
```

```python
from kss import split_sentences

splited = []
for fname in flist[:8]:
    f = open(os.path.join(dir_path, fname), 'r', encoding='utf-8')
    text = f.read()
    # print(text)
    splited.append(split_sentences(text))
    # while True:
    #     line = f.readline()
    #     if not line: break
    #     text.append(line)

    f.close()
```

Mecab으로 pos tagging후 적어도 주어, 동사가 있는 경우만 문장으로 추출할 계획

```python
from konlpy.tag import Mecab
```

```python
mecab = Mecab()
test = []
rej = []
for text in splited:
    for line in text:
        res = mecab.pos(line)
        noun_res = mecab.nouns(line) # 명사만 추출
        if line == '같이 보기': # 다른 조건들
            rej.append(line)
            continue
        flag0 = False; flag1 = False;
        for tup in res:
            if 'N' in tup[1]: flag0 = True;
            if 'V' in tup[1] or 'E' in tup[1]: flag1 = True
        if flag0 and flag1 and len(noun_res) > 1: test.append(line) # 명사와 동사가 있고 명사가 2개 이상있는 경우만
        else: rej.append(line)
```

```python
for i, line in enumerate(test):
    with open(f'./files/item{i}.txt', 'w', encoding='utf-8') as f:
        f.write(line)
```
